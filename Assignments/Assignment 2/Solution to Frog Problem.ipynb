```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
P = np.zeros((n,n))
I = np.identity(n)
R = np.ones(n)

for i in range(0,n):
    P[i,i+1:n] = 1/(n-i)
    
V = np.dot(np.linalg.inv(I-P),R)
    
print(P)
print(V)

plt.plot(np.arange(1,n+1,1),V[::-1])
plt.plot(np.arange(1,n+1,1), np.log(np.arange(1,n+1,1))+1)
plt.show()
```
